### –ó–∞–¥–∞—á–∞ 3

–°–µ–º—å —á–µ–ª–æ–≤–µ–∫ —Ä–∞—Å—Å–∞–∂–∏–≤–∞—é—Ç—Å—è –Ω–∞—É–¥–∞—á—É –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º?

#### –†–µ—à–µ–Ω–∏–µ

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–∞–∫–æ–≥–æ —Å–æ–±—ã—Ç–∏—è –±—É–¥–µ—Ç —Ä–∞–≤–Ω–∞ –æ—Ç–Ω–æ—à–µ–Ω–∏–∏—é —É—Å–ø–µ—à–Ω—ã—Ö —Å–æ–±—ã—Ç–∏–π –∫–æ –≤—Å–µ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ —Å–æ–±—ã—Ç–∏–π. –í—Å–µ–≥–æ –≤–æ–∑–º–æ–∂–Ω–æ 12 –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –æ–∫–∞–∂—É—Ç—Å—è —Ä—è–¥–æ–º –ø–ª—é—Å –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∏–∑ 5 –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –ª—é–¥–µ–π. –ê –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –≤—Å–µ—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ —Ä–∞–∑–º–µ—â–µ–Ω–∏—è 7 —á–µ–ª–æ–≤–µ–∫ –Ω–∞ —Å–∫–∞–º–µ–π–∫–µ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Ñ–æ—Ä–º—É–ª–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –∏–∑ 7 —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

```python
def permutations(n):
    return np.math.factorial(n)


m = 12 * permutations(5)
n = permutations(7)

P = m / n
```

#### –û—Ç–≤–µ—Ç

```python
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –¥–≤–∞ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã—Ö —á–µ–ª–æ–≤–µ–∫–∞ –±—É–¥—É—Ç —Å–∏–¥–µ—Ç—å —Ä—è–¥–æ–º —Ä–∞–≤–Ω–∞ {round(P*100, 2)}%')
```

### –ó–∞–¥–∞—á–∞ 5

–ë—Ä–æ—Å–∞–µ—Ç—Å—è –∏–≥—Ä–∞–ª—å–Ω–∞—è –∫–æ—Å—Ç—å. –ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ A - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á—ë—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞, —Å–æ–±—ã—Ç–∏–µ B - –ø–æ—è–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ –±–æ–ª—å—à–µ —Ç—Ä—ë—Ö. –Ø–≤–ª—è—é—Ç—Å—è –ª–∏ —ç—Ç–∏ —Å–æ–±—ã—Ç–∏—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏?

#### –†–µ—à–µ–Ω–∏–µ

–°–æ–±—ã—Ç–∏—è ùê¥ –∏ ùêµ –Ω–∞–∑—ã–≤–∞—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏, –µ—Å–ª–∏ ùëÉ(ùêµ|ùê¥)=ùëÉ(ùêµ). –ì–¥–µ P(B|A)=ùëÉ(ùê¥ùêµ)/ùëÉ(ùê¥). –í –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ P(AB) —Ä–∞–≤–æ 1/3, —Ç.–∫. —É—Å–ª–æ–≤–∏—é –≤—ã–ø–∞–¥–µ–Ω–∏—è —á–µ—Ç–Ω–æ–≥–æ —á–∏—Å–ª–∞ –±–æ–ª—å—à–µ 3 —É–¥–æ–ª–µ—Ç–≤–æ—Ä—è—é—Ç —Å–æ–±—ã—Ç–∏—è –≤—ã–ø–∞–¥–µ–Ω–∏—è —á–∏—Å–ª–∞ 4 –∏ 6. –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(A)=1/2, –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(B)=1/2.

P(B|A)=ùëÉ(ùê¥ùêµ)/ùëÉ(ùê¥) = 1/3 / 1/2 = 2/3

ùëÉ(ùêµ|ùê¥) != ùëÉ(ùêµ), —Ç.–∫. 2/3 != 1/2

#### –û—Ç–≤–µ—Ç

–°–æ–±—ã—Ç–∏—è –Ω–µ —è–≤–ª—è—é—Ç—Å—è –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏.

### –ó–∞–¥–∞—á–∞ 6

–î–æ–ø—É—Å—Ç–∏–º, –∏–º–µ–µ—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä–∞—è –æ—á–µ–Ω—å —Ä–µ–¥–∫–∞—è –±–æ–ª–µ–∑–Ω—å (–ø–æ—Ä–∞–∂–∞–µ—Ç 0.1% –Ω–∞—Å–µ–ª–µ–Ω–∏—è). –í—ã –ø—Ä–∏—Ö–æ–¥–∏—Ç–µ –∫ –≤—Ä–∞—á—É, –≤–∞–º –¥–µ–ª–∞—é—Ç —Ç–µ—Å—Ç –Ω–∞ —ç—Ç—É –±–æ–ª–µ–∑–Ω—å, –∏ —Ç–µ—Å—Ç –æ–∫–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º. –í—Ä–∞—á –≥–æ–≤–æ—Ä–∏—Ç –≤–∞–º, —á—Ç–æ —ç—Ç–æ—Ç —Ç–µ—Å—Ç –≤–µ—Ä–Ω–æ –≤—ã—è–≤–ª—è–µ—Ç 99 % –±–æ–ª—å–Ω—ã—Ö —ç—Ç–æ–π –±–æ–ª–µ–∑–Ω—å—é –∏ –≤—Å–µ–≥–æ –ª–∏—à—å –≤ 1% —Å–ª—É—á–∞–µ–≤ –¥–∞—ë—Ç –ª–æ–∂–Ω—ã–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç.

–í–æ–ø—Ä–æ—Å: –≤–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ –≤—ã –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª—å–Ω—ã –µ–π?

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–æ—Ä–º—É–ª—É –ë–∞–π–µ—Å–∞ —Å —Ä–∞—Å–∫—Ä—ã—Ç–∏–µ–º –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—è —Å –ø–æ–º–æ—â—å—é —Ñ–æ—Ä–º—É–ª—ã –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏.

#### –†–µ—à–µ–Ω–∏–µ

–ü—É—Å—Ç—å —Å–æ–±—ã—Ç–∏–µ A - —Ç–µ—Å—Ç –æ–∫–∞–∑–∞–ª—Å—è –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–º. –°–æ–±—ã—Ç–∏–µ B - —è –æ–∫–∞–∑–∞–ª—Å—è –±–æ–ª–µ–Ω. –¢–æ–≥–¥–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —è –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª–µ–Ω –ø—Ä–∏ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ —Ç–µ—Å—Ç–∞ –±—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ë–∞–π–µ—Å–∞.

P(B|A) = (P(A|B) * P(B)) / P(A)

–≥–¥–µ:
P(A|B) —ç—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–≥–æ —Ç–µ—Å—Ç–∞ –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ —è –±–æ–ª–µ–Ω, –æ–Ω–∞ —Ä–∞–≤–Ω–∞ 0.99 –ø–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏
P(B) —Ä–∞–≤–Ω–∞ 0.001 —Ç–∞–∫ –∂–µ –ø–æ —É—Å–ª–æ–≤–∏—é –∑–∞–¥–∞—á–∏
P(A) —Ä–∞—Å–ø–∏—à–µ–º –ø–æ —Ñ–æ—Ä–º—É–ª–µ –ø–æ–ª–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏. –£ –Ω–∞—Å –µ—Å—Ç—å –¥–≤–µ –≥–∏–ø–æ—Ç–µ–∑—ã: H1 - —Ç–µ—Å—Ç –≤–µ—Ä–Ω–æ –ø–æ–∫–∞–∑–∞–ª —Ä–µ–∑—É–ª—å—Ç–∞—Ç, H2 - —Ç–µ—Å—Ç –ª–æ–∂–Ω–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–π.

–¢–æ–≥–¥–∞ P(A) = P(H1) * P(A|H1) + P(H2) * P(A|H2).

–ì–¥–µ P(H1) = 0.99, P(H2) = 0.01, P(A|H1) = 0.001 (0.1% –Ω–∞—Å–µ–ª–µ–Ω–∏—è), P(A|H2) = 0.999 (99.9% –Ω–∞—Å–µ–ª–µ–Ω–∏—è)

–ü–æ–¥—Å—Ç–∞–≤–∏–º –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ñ–æ—Ä–º—É–ª—É:

P(B|A) = (0.99 * 0.001) / (0.99 * 0.001 + 0.01 * 0.999)


```python
P_ba = (0.99 * 0.001) / (0.99 * 0.001 + 0.01 * 0.999)
```

#### –û—Ç–≤–µ—Ç

```python
print(f'–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —è –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –±–æ–ª–µ–Ω —Ä–∞–≤–Ω–∞ {round(P_ba*100, 2)}%')
```
